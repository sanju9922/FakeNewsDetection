```python


import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import word_tokenize
from scipy.stats import ttest_ind
import graphviz
```

```python
# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
# Read the data
data_fake = pd.read_csv("Fake.csv")
data_true = pd.read_csv('True.csv')
```

```python
# Assign class labels
data_fake["class"] = 0
data_true["class"] = 1
```

```python
# Combine the datasets
news = pd.concat([data_fake, data_true], axis=0)
news = news.drop(['title', 'subject', 'date'], axis=1)
news = news.sample(frac=1).reset_index(drop=True)
```

```python
# Preprocessing function
def wordopt(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d', '', text)
    text = re.sub(r'\n', ' ', text)
    
    # Tokenization
    words = word_tokenize(text)
    
    # Remove stopwords
    words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    
    # Lemmatization and stemming
    lemmatizer = WordNetLemmatizer()
    stemmer = PorterStemmer()
    words = [lemmatizer.lemmatize(word) for word in words]
    words = [stemmer.stem(word) for word in words]
    
    return ' '.join(words)
```

```python
# Apply preprocessing
news['text'] = news['text'].apply(wordopt)
```

```python
# Split data into features and labels
x = news['text']
y = news['class']
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Vectorize text data
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
# Train and evaluate models
def train_and_evaluate_model(model, name):
    model.fit(xv_train, y_train)
    predictions = model.predict(xv_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{name}")
    print(classification_report(y_test, predictions))
    return accuracy, confusion_matrix(y_test, predictions)
```

```python
# Logistic Regression
LR = LogisticRegression()
lr_accuracy, lr_cm = train_and_evaluate_model(LR, "Logistic Regression")
```

```python
# Decision Tree Classifier
DTC = DecisionTreeClassifier()
dtc_accuracy, dtc_cm = train_and_evaluate_model(DTC, "Decision Tree Classifier")
```

```python
# Random Forest Classifier
rfc = RandomForestClassifier()
rfc_accuracy, rfc_cm = train_and_evaluate_model(rfc, "Random Forest Classifier")
```

```python
# Gradient Boosting Classifier
gbc = GradientBoostingClassifier()
gbc_accuracy, gbc_cm = train_and_evaluate_model(gbc, "Gradient Boosting Classifier")
```

```python
accuracy_scores = {
    'Logistic Regression': lr_accuracy,
    'Decision Tree': dtc_accuracy,
    'Random Forest': rfc_accuracy,
    'Gradient Boosting': gbc_accuracy
}
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=list(accuracy_scores.keys()), y=list(accuracy_scores.values()))
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Classifiers')
plt.show()
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Fake', 'Real'], yticklabels=['Fake', 'Real'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
```

```python
# Plot confusion matrices
plot_confusion_matrix(lr_cm, "Confusion Matrix for Logistic Regression")
plot_confusion_matrix(dtc_cm, "Confusion Matrix for Decision Tree")
plot_confusion_matrix(rfc_cm, "Confusion Matrix for Random Forest")
plot_confusion_matrix(gbc_cm, "Confusion Matrix for Gradient Boosting")
```
